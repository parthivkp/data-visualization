```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') #default dataset seaborn
```

```python
tips.head()
```

```python
sns.distplot(tips["total_bill"])
sns.set_theme(style="darkgrid")
#give kde="false to remove that line ..give as parameter in distplot"
##these are distribution plots
```

```python
#joint plots to compare distribution for 2 variables
sns.jointplot(x="total_bill",y="tip",data=tips,hue="sex")
```

```python
#all variable distribution, extra parameter hue="sex" for catergorical data like male and female
sns.pairplot(tips,hue="sex")
```

```python
#scatter points using hex points
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex") 
```

```python

#### these are categorial plots like sex,day etc......


sns.barplot(x="sex",y="total_bill",data=tips) #estimator=mean default 
# sns.barplot(x="sex",y="total_bill",data=tips,estimator=)
```

```python
sns.countplot(tips["sex"])
# or sns.countplot(x="sex",data=tips)
#counts number of male and femaile in data set
```

```python
sns.countplot(tips["day"])
#counts number people visiting each day
```

```python
#box and wisker plot
fig=plt.figure(figsize=(10,7))
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
fig=plt.figure(figsize=(10,5))
sns.violinplot(x="day",y="total_bill",data=tips,hue="sex",split=True)
#split for combining male and female
#almost same as box plot
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,hue="smoker")
#like scatter plot
#if its in a straight line and number of dots  arent visible make jitter=True as parameter
```

```python

sns.swarmplot(x="day",y="total_bill",data=tips)
#sometimes it not suitable for large data set....cant scale properly (sometimes)
```

```python
sns.violinplot(x="day",y="total_bill",data=tips)
sns.swarmplot(x="day",y="total_bill",data=tips,color="black")
#combine with voilin plot
```
